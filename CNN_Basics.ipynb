```python
from ast import increment_lineno
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train, y_train ), (X_test, y_test)= keras.datasets.mnist.load_data()
```

```python
X_train.shape
```

```python
y_train[0]
```

```python
y_train[:5]
```

```python
X_train_flattened=X_train.reshape(60000, 784)
```

```python
X_train_flattened.shape
```

```python
X_test_flattened=X_test.reshape(len(X_test), 28*28)
X_test_flattened.shape
```

```python
model = keras.Sequential()

model.add(keras.layers.Input((784,0)))

model.add(keras.layers.Dense(units= 10, activation='sigmoid'))


keras.utils.plot_model(model, "my_first_model.png", show_shapes=True)
```

```python



model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation='sigmoid') # Added assignment to input_shape

])
model.compile(
     optimizer='adam',
     loss='sparse_categorical_crossentropy',
     metrics=['accuracy'] # Fixed typo in metrics
)

history = model.fit(X_train_flattened, y_train, epochs=5)
```

```python
# plot history of model loss and accuracy
plt.plot(history.history['loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```

```python
# plot history of model accuracy
plt.plot(history.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()
```
